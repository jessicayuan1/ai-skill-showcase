#Step 1: Create Two Images

```python
#Importing required Libraries
from PIL import Image, ImageDraw, ImageFilter, ImageOps
import numpy as np
import matplotlib.pyplot as plt

# Creates two images

W= 256 #defining width and heiight
H= 256
imgA = Image.new('RGB', (W,H), 0) #creates a black background image
object1 = ImageDraw.Draw(imgA) #creates a drawing object to place on background
object1.ellipse((10, 10, 246, 246), fill=(255, 0, 0))       # red
object1.ellipse((26, 26, 230, 230), fill=(255, 127, 0))     # orange
object1.ellipse((42, 42, 214, 214), fill=(255, 255, 0))     # yellow
object1.ellipse((58, 58, 198, 198), fill=(0, 255, 0))       # green
object1.ellipse((74, 74, 182, 182), fill=(0, 0, 255))       # blue
object1.ellipse((90, 90, 166, 166), fill=(75, 0, 130))      # indigo
object1.ellipse((106, 106, 150, 150), fill=(148, 0, 211))   # violet





imgB = Image.new('RGB', (W,H), 0) #creates a black background image
object2 = ImageDraw.Draw(imgB) #creates a drawing object to place on background
object2.ellipse((78,38,178,198), fill=(255, 255, 255)) #draws a white oval
object2.ellipse((100, 80, 115, 95), fill=(0, 0, 0))   # left eye
object2.ellipse((140, 80, 155, 95), fill=(0, 0, 0))   # right eye
object2.arc((90, 130, 165, 170), start=0, end=180, fill=(0, 0, 0), width=8)  # smile

object2.polygon([(100, 100), (107, 120), (115, 100)], fill=(128,0,128)) #marking the face with purple triangles
object2.polygon([(140, 100), (147, 120), (155, 100)], fill=(128,0,128))
object2.polygon([(100, 75), (107, 55), (115, 75)], fill=(128,0,128))
object2.polygon([(140, 75), (147, 55), (155, 75)], fill=(128,0,128))

#Shows the images using matplotlib
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(imgA, cmap='gray')
plt.title("Image A")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(imgB, cmap='gray')
plt.title("Image B")
plt.axis('off')
plt.show()

```

#Step 2: Apply a transformation (Blur and Inverse + Border)

```python

blurA = imgA.filter(ImageFilter.GaussianBlur(3)) #applies gaussian blur to image A
invertedB = ImageOps.invert(imgB) #inverts image B

invertedB = ImageOps.expand(invertedB, border=5, fill='black') #adds a black border to inverted image to make it look nicer
```

#Step 3: Display Transformed Images

```python
#Shows the images using matplotlib
plt.figure(figsize=(16,8))
plt.subplot(1,4,1)
plt.imshow(imgA, cmap='gray')
plt.title("Image A")
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(imgB, cmap='gray')
plt.title("Image B")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(blurA, cmap='gray')
plt.title("Image A Transformed")
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(invertedB, cmap='gray')
plt.title("Image B Transformed")
plt.axis('off')
plt.show()
```

#Step 4: Method of Aproach
##Image Generation
    - Image 1: Created a rainbow in a cirlce using ellipses and RGB values
    - Image 2: created No-Face from spirited away using simple geometry

##Transforamtions Applied
    - Image 1: Applied Gaussian blur to smooth the change in colour creating a gradient look
    - Image 2: inverted No-Face's colour pallete
